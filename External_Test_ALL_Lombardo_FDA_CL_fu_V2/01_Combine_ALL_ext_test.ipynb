```python
from rdkit.Chem import inchi
from rdkit import Chem
from pandarallel import pandarallel
pandarallel.initialize()
from molvs import Standardizer
s = Standardizer()

def standardized_nonisomeric_smiles_from_smiles(value):

    try: return Chem.MolToSmiles(s.standardize(Chem.MolFromSmiles(value)), isomericSmiles=False)
    except: return "Cannot do"
    
def smilesr_from_smiles(value):

    try: return Chem.MolToSmiles(s.standardize(Chem.MolFromSmiles(value)))
    except: return "Cannot do"
    
def standardised_inchi_from_smile(value):

    try: return Chem.MolToInchi(s.standardize(Chem.MolFromSmiles(value)))
    except: return "Cannot_do"
    

lombardo["smiles_r"] = lombardo["smiles_r"].parallel_apply(smilesr_from_smiles)  
lombardo["InChICode_standardised"] = lombardo["smiles_r"].parallel_apply(standardised_inchi_from_smile)  
lombardo["nonisomeric_smiles_r"] = lombardo["smiles_r"].parallel_apply(standardized_nonisomeric_smiles_from_smiles)   
lombardo = lombardo[lombardo["smiles_r"]!="Cannot_do"]   

CL["smiles_r"] = CL["smiles_r"].parallel_apply(smilesr_from_smiles)  
CL["InChICode_standardised"] = CL["smiles_r"].parallel_apply(standardised_inchi_from_smile)      
CL["nonisomeric_smiles_r"] = CL["smiles_r"].parallel_apply(standardized_nonisomeric_smiles_from_smiles) 
CL = CL[CL["smiles_r"]!="Cannot_do"]  

FDA["smiles_r"] = FDA["smiles"].parallel_apply(smilesr_from_smiles)  
FDA["InChICode_standardised"] = FDA["smiles_r"].parallel_apply(standardised_inchi_from_smile)      
FDA["nonisomeric_smiles_r"] = FDA["smiles_r"].parallel_apply(standardized_nonisomeric_smiles_from_smiles) 
FDA = FDA[FDA["smiles_r"]!="Cannot_do"]  

fu["smiles_r"] = fu["smiles_r"].parallel_apply(smilesr_from_smiles)  
fu["InChICode_standardised"] = fu["smiles_r"].parallel_apply(standardised_inchi_from_smile)      
fu["nonisomeric_smiles_r"] = fu["smiles_r"].parallel_apply(standardized_nonisomeric_smiles_from_smiles) 
fu = fu[fu["smiles_r"]!="Cannot_do"]  
```

```python
from dimorphite_dl.dimorphite_dl import DimorphiteDL
from rdkit.Chem import AddHs
from rdkit.Chem.MolStandardize import rdMolStandardize

def standardize(smiles):
    # follows the steps in
    # https://github.com/greglandrum/RSC_OpenScience_Standardization_202104/blob/main/MolStandardize%20pieces.ipynb
    # as described **excellently** (by Greg) in
    # https://www.youtube.com/watch?v=eWTApNX8dJQ
    try: 
        mol = Chem.MolFromSmiles(smiles)
        #print(smiles)
        
        # removeHs, disconnect metal atoms, normalize the molecule, reionize the molecule
        clean_mol = rdMolStandardize.Cleanup(mol) 
        #print(Chem.MolToSmiles(clean_mol))
        
        # if many fragments, get the "parent" (the actual mol we are interested in) 
        parent_clean_mol = rdMolStandardize.FragmentParent(clean_mol)

        # try to neutralize molecule
        uncharger = rdMolStandardize.Uncharger() # annoying, but necessary as no convenience method exists
        uncharged_parent_clean_mol = uncharger.uncharge(parent_clean_mol)

        #print(uncharged_parent_clean_mol)
        
        # use pH 7.4 https://git.durrantlab.pitt.edu/jdurrant/dimorphite_dl/
        dimorphite = DimorphiteDL(min_ph=7.4, max_ph=7.4, pka_precision=0)
        protonated_smiles = dimorphite.protonate(Chem.MolToSmiles(uncharged_parent_clean_mol))

        #print("protonated_smiles")
        
        if len(protonated_smiles) > 0:
                protonated_smile = protonated_smiles[0]

        protonated_mol = Chem.MolFromSmiles(protonated_smile)
        #protonated_mol= AddHs(protonated_mol)
        #protonated_smile = Chem.MolToSmiles(protonated_mol)


        # attempt is made at reionization at this step
        # at 7.4 pH

        te = rdMolStandardize.TautomerEnumerator() # idem
        taut_uncharged_parent_clean_mol = te.Canonicalize(protonated_mol)
     
        return  Chem.MolToSmiles(taut_uncharged_parent_clean_mol)
    
    except: 
        
        return "Cannot_do"

def inchi_from_standardised_smile(value):

    try: return Chem.MolToInchi(Chem.MolFromSmiles(value))
    except: return "Cannot_do"
```

```python
ext_test_all["smiles_r"] = ext_test_all["smiles_r"].parallel_apply(standardize)  
ext_test_all = ext_test_all[ext_test_all["smiles_r"]!="Cannot_do"].reset_index(drop=True)
ext_test_all
```
